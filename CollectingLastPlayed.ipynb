## Collecting the Date Last Played

We are collecting this so that I can make claims about how many players are active

```python
from halo import halo
import json
```

```python
halo_class = halo('LastPlayedData.csv')
```

```python
# removes already scanned ones
gamertag = halo_class.gamertags(nasort='Yes')

for i,itm in enumerate(gamertag):
    try:
        out = halo_class.getLastPlayed(itm)
        print(i,out)
    except KeyboardInterrupt:
        break
    #These appear to be the errors when they occur
    # I just skip over them
    except json.JSONDecodeError:
        print('Error JSON')
    except IndexError:
        print('Error Index')
```
